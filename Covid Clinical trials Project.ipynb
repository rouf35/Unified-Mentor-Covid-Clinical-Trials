

                                                          Univariate Analysis
                             ● Status Distribution: Analyze the status of clinical trials (e.g., Completed, Ongoing).

                                                                   

```python
# Distribution of Clinical Trial Statuses

status_counts = data["Status"].value_counts()
plt.figure(figsize=(12, 6))
status_counts.plot(kind="bar", color="cyan", edgecolor="black")
plt.xlabel("Clinical Trial Status")
plt.ylabel("Count")
plt.title("Distribution of Clinical Trial Statuses")

plt.show()
```


                                        ● Phase Distribution: Understand the distribution of trial phases.

                                        

```python
#Distribution of Clinical Trial Phases

phase_counts = data["Phases"].value_counts()
plt.figure(figsize=(12, 6))
phase_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.xlabel("Clinical Trial Phase")
plt.ylabel("Count")
plt.title("Distribution of Clinical Trial Phases")
```


                                      
                                      ● Age Group Analysis: Analyze the distribution of age groups.

                                      

```python
#● Age Group Analysis: Analyze the distribution of age groups.
# For better Analysis I have grouped them in three age groups as mentioned below
age_counts = data["Age"].value_counts()
child_keywords = ["Child"]
adult_keywords = ["Adult"]
older_adult_keywords = ["Older Adult"]
child_count = 0
adult_count = 0
older_adult_count = 0

for age_group, count in age_counts.items():
    if any(keyword in age_group for keyword in child_keywords):
        child_count += count
    if any(keyword in age_group for keyword in adult_keywords):
        adult_count += count
    if any(keyword in age_group for keyword in older_adult_keywords):
        older_adult_count += count
        
age_grouped_counts = {
    "Child (0-17 years)": child_count,
    "Adult (18-64 years)": adult_count,
    "Older Adult (65+ years)": older_adult_count,
}
plt.figure(figsize=(12, 6))
plt.bar(age_grouped_counts.keys(), age_grouped_counts.values(), color=["royalblue", "seagreen", "tomato"], edgecolor="black")

plt.xlabel("Age Group")
plt.ylabel("Count")
plt.title("Grouped Distribution of Age Groups in Clinical Trials")

plt.show()
```


                                                                
                                                                Bivariate Analysis
                                                Explore relationships between different variables.
                                    ● Status vs. Phases: Explore how trial phases are distributed across different statuses.

                                    

```python
# Distribution of Trial Phases Across Different Statuses
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x="Phases", hue="Status", order=data["Phases"].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel("Trial Phase")
plt.ylabel("Count")
plt.title("Distribution of Trial Phases Across Different Statuses")
plt.legend(title="Status")
plt.show()
```


                                              
                                              7. Time Series Analysis
                        Analyze the trends over time, such as the number of trials started over the months.

                        

```python
data['Start Date'] = pd.to_datetime(data['Start Date'],errors='coerce')
data['Completion Date'] = pd.to_datetime(data['Completion Date'],errors='coerce')
```

```python
data['Month'] = data['Start Date'].dt.to_period('M')  # Extract year-month
trials_per_month = data.groupby('Month').size()
```

```python
plt.figure(figsize=(10, 5))
trials_per_month.plot(marker='o', linestyle='-')
plt.xlabel("Month")
plt.ylabel("Number of Trials Started")
plt.title("Clinical Trials Started Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
